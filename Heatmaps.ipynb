```python
import numpy as np
import seaborn as sns
import heatmaps as h
import matplotlib.pyplot as plt
```

```python
def plot_LH(vals1, vals2, ind_var_1, ind_var_2, filename = None):
    if filename is None:
        filename = f"{ind_var1}_{ind_var2}.npy"
    yy = np.load(filename)
    
    h = len(vals1)
    w = len(vals2)
    
    fig, axs = plt.subplots(h, w, figsize= (16,16))
    yy = yy.reshape(h,w,10000)
    yy = np.flip(yy,0)
    
    tt = np.linspace(0,128,10000)
    
    vals1 = vals1[::-1]
    
    for i in range(h):
        for j in range(w):
            axs[i,j].plot(tt, yy[i,j])
            axs[i,j].set_ylim([0, 150])
            if i == h-1:
                axs[i,j].set_xlabel(f'{ind_var_2} = {vals2[j]}')
            if j == 0:
                axs[i,j].set_ylabel(f'{ind_var_1} = {vals1[i]}')
```

# Trying out FFT instead of PSD

```python
from scipy.fft import fft, fftfreq
# Number of sample points
vals1 = np.linspace(0, 92, 6)
vals2 = np.linspace(0, 1.3, 6)
ind_var1 = "e_dose"
ind_var2 = "p_dose"

time_series = np.load(f"{ind_var1}_{ind_var2}.npy")
time_series.shape

time_series = time_series.reshape((6,6,10000))

N = 10000
T = 128/10000
yf = fft(time_series[0,0])
xf = fftfreq(N, T)[:N//2]
import matplotlib.pyplot as plt
plt.plot(xf[1:], 2.0/N * np.abs(yf[1:N//2]))
plt.grid()
plt.show()

idx = np.argmax(yf[1:])
xf[1:][idx]
```

```python
up_lambda_shift = lambda lam_up : np.log((1-0.05)/0.05)/lam_up
down_lambda_shift = lambda lam_down : -np.log(0.95/(1-0.95))/lam_down
```

```python
up_lambda_shift(1)
```

```python
lambdas = np.linspace(0.5,10,21)
```

```python
time_to_95 = 2*np.array(list(map(up_lambda_shift, lambdas)))
```

```python
one_line = np.ones(21)
plt.plot(lambdas, time_to_95)
plt.plot(lambdas, one_line, '.', label = 'Days = 1')
plt.legend()
plt.title('Time to reach 95% of full dose')
plt.ylabel('Days')
plt.xlabel('Lambda')
```
