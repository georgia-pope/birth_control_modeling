# Trying out FFT instead of PSD

```python
from scipy.fft import fft, fftfreq
# Number of sample points
vals1 = np.linspace(0, 92, 6)
vals2 = np.linspace(0, 1.3, 6)
ind_var1 = "e_dose"
ind_var2 = "p_dose"

time_series = np.load(f"{ind_var1}_{ind_var2}.npy")
time_series.shape

time_series = time_series.reshape((6,6,10000))

N = 10000
T = 128/10000
yf = fft(time_series[0,0])
xf = fftfreq(N, T)[:N//2]
import matplotlib.pyplot as plt
plt.plot(xf[1:], 2.0/N * np.abs(yf[1:N//2]))
plt.grid()
plt.show()

idx = np.argmax(yf[1:])
xf[1:][idx]
```
