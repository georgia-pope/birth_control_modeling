# Normal Cycle

```python
variables, matrix = j.create_matrix(0,0)
```

```python
matrix
```

```python
jacobian = matrix.jacobian(variables)
```

```python
jacobian
```

```python
RP_LH, LH, RP_FSH, FSH, RcF, GrF, DomF, Sc_1, Sc_2, Lut_1, Lut_2, Lut_3, Lut_4 = variables


subs = {
    RP_LH:29.65,
    LH:6.86,
    RP_FSH: 8.47,
    FSH:6.15,
    RcF:3.83,
    GrF:11.51,
    DomF:5.48,
    Sc_1:19.27,
    Sc_2: 45.64,
    Lut_1: 100.73,
    Lut_2: 125.95,
    Lut_3: 135.84,
    Lut_4: 168.71
}

evaled_jacobian = jacobian.evalf(subs=subs)

evaled_jacobian
```

```python
np_jacobian = np.array(evaled_jacobian.tolist()).astype(np.float64)
```

```python
import numpy as np
eig_val, eig_vec = np.linalg.eig(np_jacobian)

eig_val
```
